# Galon Başına Mil (Miles per Gallon)

Bu veri seti kapsamında Almanya, Japonya ve Amerika Birleşik Devletleri tarafından üretilen arabaların yakıt tüketi başta olmak üzere, diğer özelliklerini de görselleştirerek kıyaslayacağız.

## Veri Seti

İlgili veri setine buradan ulaşılabilir: https://raw.githubusercontent.com/cads-tedu/DSPG/master/Veri%20Setleri/mpg.csv

## Değişkenler (Özellikler, Kolonlar, (İngilizce: Variables, Features, Columns))

**manufacturer:** Üretici, arabanın markası

**model:** Arabanın modeli

**displ:** Silindir hacmi

**year:** Üretim yılı

**cyl:** Silindir sayısı

**trans:** Şanzıman tipi

**drv:** Çekiş tipi (önden ya da arkadan çekişli)

**cty:** Şehir içi tüketilen galon başına mil

**hwy:** Şehir dışı tüketilen galon başına mil

**fl:** Aracın kullandığı yakıt tipi

**class:** Araç tipi
